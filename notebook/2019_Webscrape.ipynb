```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.chrome.options import Options 
import chromedriver_autoinstaller
```

Create a function that opens up a proxy webdriver. Access the link(https://fantasydata.com/nfl/advanced-rb-metrics) from the driver. I'm gonna start with 2019 and move backwards. 

```python
 #Create a Function that initializes a Webdriver
def create_driver():
    driver = chromedriver_autoinstaller.install(cwd=True)
    driver = webdriver.Chrome(driver)
    return driver
```

```python
driver = create_driver()
```
